## Defining the business problem to be solved

Spotify is one of the largest music streaming platforms in the world, with over 600 million active users and a catalogue of over 100 million tracks. Spotify has identified an opportunity to improve their music cataloguing and metadata systems by using AI/ML models to predict the release era of a song based on its audio features. Incorrectly dated or undated tracks cause poor playlist curation, recommendation errors, and metadata inconsistencies. Spotify has approached you as a Data Engineer to develop a PoC release year prediction service that estimates the album release year of a track based on measurable audio features such as energy, danceability, loudness, and tempo.

## Refactoring the business problem into a machine learning problem

Given a set of audio features for a Spotify track (such as energy, danceability, acousticness, loudness, tempo, and speechiness), predict the year the track's album was released. This is a supervised regression problem where the target variable is `track_album_release_date` (as a year), and the input features are numerical audio characteristics extracted by Spotify's internal ML pipeline.

## Defining success metrics

- The model should be able to predict the release year of a track within ±5 years for at least 70% of predictions, reflecting practical usefulness for metadata correction.
- The model should outperform a naive baseline (e.g. always predicting the mean release year) by a meaningful margin.
- The service should be reliable enough to flag tracks with likely incorrect or missing release year metadata for human review.
- Success is defined as reducing metadata inconsistencies in Spotify's catalogue, improving playlist curation accuracy and recommendation quality.

## Researching available data

Spotify have sourced a validated raw dataset of 32,833 tracks with playlist and audio feature metadata. The data is saved in `spotify_songs.csv`.

- The data columns are:

| Column | Data | Datatype | Range | Units |
| ------ | ---- | -------- | ----- | ----- |
| track_popularity | Measure of track popularity | float | 0 to 100 | Score |
| track_album_release_date | Release year of album **(target)** | int | ~1957–2020 | Year |
| playlist_genre | Genre of playlist | string | N/A | N/A |
| playlist_subgenre | Subgenre of playlist | string | N/A | N/A |
| danceability | How suitable a track is for dancing | float | 0 to 1 | Arbitrary |
| energy | Perceptual intensity and activity | float | 0 to 1 | Arbitrary |
| key | Musical key of the track | int | 0 to 11 | Pitch class |
| loudness | Overall loudness of the track | float | -60 to 0 | dB |
| mode | Modality (1=major, 0=minor) | bool | 0 or 1 | N/A |
| speechiness | Presence of spoken words | float | 0 to 1 | Arbitrary |
| acousticness | Confidence track is acoustic | float | 0 to 1 | Arbitrary |
| instrumentalness | Probability of no vocals | float | 0 to 1 | Arbitrary |
| liveness | Probability of live performance | float | 0 to 1 | Arbitrary |
| valence | Musical positiveness | float | 0 to 1 | Arbitrary |
| tempo | Estimated tempo | float | ~30–250 | BPM |
| duration_ms | Duration of track | float | ~29500–517800 | Milliseconds |
